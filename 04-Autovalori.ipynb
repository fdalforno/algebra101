# Autovalori

Abbiamo visto nelle lezioni precedenti che una matrice può rappresentare una trasformazione lineare.
Prendiamo ad esempio la matrice:

$$\begin{bmatrix} 3 & -2 \\ 1 & 0 \end{bmatrix}$$

Se applichiamo questa matrice di trasformazione al vettore 

$$\begin{bmatrix} -1 \\ 1  \end{bmatrix}$$

otteniamo come risultato:

$$\begin{bmatrix} -5 \\ -1  \end{bmatrix}$$

Abbiamo un cambiamento di lunghezza e direzione
Ora proviamo ad applicare la trasformazione al vettore 

$$\begin{bmatrix} 2 \\ 1  \end{bmatrix}$$

Otteniamo il risultato:

$$\begin{bmatrix} 4 \\ 2  \end{bmatrix}$$
Che cosa è successo? il vettore risultante non ha cambiato direzione ma solo la sua lunghezza.
Ecco possiamo dire che per la matrice di trasformazione A un vettore che ha la caratteristica descritta sopra si può chiamare autovettore.

### Autovettore
Definiamo un  autovettore un vettore che applicato ad una matrice di trasformazione risulta un multiplo scalare di se stesso

### Autovalore
L'autovalore rappresenta il fattore di scala descritto prima

## Calcolo degli autovalori

Date le seguenti definizioni:

* $A$ matrice di trasformazione
* $\overline{X}$ rappresenta un autovettore 
* $\lambda$ rappresenta un autoalore

Gli autovettori della matrice $A$ rispettano la seguente ugualianza:

$$A \cdot \overline{X} = \lambda \cdot \overline{X}$$


Introduciamo un nuovo simbolo $I$ che rappresenta la matrice identità la seguente equazione è ancora valida:

$$A \cdot \overline{X} = I \cdot \lambda \cdot \overline{X} c  \lambda \cdot I \cdot \overline{X}$$

Raccogliamo l'equazione e troviamo che:

$$(A - \lambda I) \cdot \overline{X} = 0$$

Visto che non vogliamo trovare le soluzioni banali di $\vec{v}$ l'unico sistema per ottenere una soluzione del genere è che il determinante della matrice sia 0


$\det(A - \lambda I) = 0$

![determinant](./images/eigenvalue.png)


Facciamo un esempio pratico per aiutarci calcoliamo gli autovalori della matrice:

$$\begin{bmatrix} 2 && 3 \\ 3 && -6  \end{bmatrix}$$

creiamo la matrice per poi derivare l'equazione caratteristica:

$$A - \lambda I = \begin{bmatrix} 2 && 3 \\ 3 && -6  \end{bmatrix} - \begin{bmatrix} \lambda && 0 \\ 0 && \lambda  \end{bmatrix} \rightarrow \begin{bmatrix} 2 - \lambda && 3 \\ 3 && -6 - \lambda  \end{bmatrix}$$

per trovare gli autovalori dobbiamo calcolare:

$$\det \left( \begin{bmatrix} 2 - \lambda && 3 \\ 3 && -6 - \lambda  \end{bmatrix} \right) = 0$$

da questa otteniamo il polinomio:

$$\lambda^{2} + 4\lambda - 21 = 0$$

Le cui soluzioni sono 3 e -7
Ora che abbiamo gli autovalori possiamo calcolare gli autovettori applicato il valore lambda otteniamo un sistema di equazioni che una volta risolta ci da il vettore o l'insieme di vettori associati.

Facciamo un esempio pratico, prendiamo la martice:

$$\begin{bmatrix} 5 && 6 \\ 3 && -2  \end{bmatrix}$$

che ha come autovalori i seguenti numeri : $\lambda_1 = 7$ e $\lambda_2 = -4$.
Partiamo dall'autovalore 7 e otteniamo il seguente sistema:

$$\begin{bmatrix} -2 && 6 \\ 3 && -2  \end{bmatrix} \begin{bmatrix} X_1 \\ X_2  \end{bmatrix} = 0$$

le due righe sono linearmente dipendenti perciò abbiamo soluzioni infinite. Tutte descritte dall'equazione:

$$X_1 - 3X_2 = 0$$

Concettualmente questa linea rappresenta l'asse di rotazione, su questo asse l'operazione non cambia la direzione del vettore.
