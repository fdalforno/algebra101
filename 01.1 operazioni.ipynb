# Operazioni sui vettori


## Prodotto scalare 
Abbiamo visto prima che numericamente parlando il prodotto scalare di due vettori aventi la stessa dimensione si calcola nel seguente modo:

$$\begin{bmatrix} a \\ b \\ c \end{bmatrix} \cdot \begin{bmatrix} e \\ f \\ g \end{bmatrix} = a \times e +  b \times f + c \times g$$

abbiamo poi visto geometricamente che il prodotto vettoriale rappresenta la proiezione di un vettore sull'altro moltiplicato per la lunghezza del secondo vettore, come si può vedere dall'immagine:


![dot product](images/codt.png)

se i vettori puntano in direzioni opposte il risultato sarà negativo, se puntano nella stessa direzione il risultato sarà un valore maggiore di zero mentre se sono perpendicolari il risultato sarà 0.
 
Ora in che modo se moltiplichiamo i vettori $\vec{v}$ e $\vec{w}$ oppure $\vec{w}$ e $\vec{v}$ otterremo lo stesso risultato.

Partiamo da un caso semplice se $\vec{v}$ e $\vec{w}$ hanno la stessa lunghezza andremo a generare una simmetria

![simmetry](images/simmetry.png)

La proiezione di  $\vec{v}$ su $\vec{w}$ sarà la stessa del contrario, ora se allunghiamo del doppio il vettore $\vec{v}$ e lo portiamo a $2 \vec{v}$ la simmetria viene rotta ma la moltiplicazione seguirà la nostra ugualianza:

$(2 \vec{v}) \cdot \vec{w} = 2(\vec{v} \cdot \vec{w})$

Se proiettiamo $\vec{w}$ su $\vec{v}$ la proiezione sarà uguale alla precedente ma il vettore $\vec{v}$ è il doppio di prima, mentre se $\vec{v}$ su $\vec{w}$ la proiezione sarà il doppio di prima ma il vettore $\vec{w}$ rimane uguale.

Dunque anche se andiamo a rompere la simmetria l'effetto della scalatura si ribalta in modo uguale su entrambe le interpretazioni


Ma come possiamo unire la visione numerica con la visione geometrica del prodotto scalare?
Qui ci viene in contro il concetto di dualità.

Il prodotto scalare può essere visto come una trasformazione lineare che mappa un vettore in più dimensioni in un numero reale, una retta.

Visto che questa operazione è lineare devono valere le seguenti equazioni:

$$L(\vec{v} + \vec{w}) = L(\vec{v}) +  L(\vec{w})$$
$$L(c\vec{v}) = c \cdot L(\vec{v})$$

Ne risulta che se applichiamo una trasformazione ad una serie di punti su una linea retta equidistanziati, questi dopo la trasformazione saranno ancora equidistanziati.

# Determinante di una matrice

Prima di parlare degli autovettori e autovalori vediamo il concetto grafico del determinante di una matrice.
Abbiamo visto che una matrice può essere vista come una funzione che trasforma un vettore in un altro.

Ma se ragioniamo su una area quale sarà l'effetto di queste funzioni?
Per esempio prendiamo la matrice di trasformazione:

$$
\begin{bmatrix} 3 & 0 \\ 0 & 2 \end{bmatrix} 
$$

questo scala il vettore $i$ di un fattore 2 e il vettore $j$ di un fattore 3.
Se prendiamo l'area coperta dai vettori $i$ e $j$ (che ha come valore 1) quale sarà il risultato della trasformazione?

Se vediamo ora l'area ha un valore pari a 6
Questo valore di "scalatura" che cambia ogni singola area viene chiamato **determinante**.
Ora se il determinate di una matrice è Zero, vuol dire che o una riga e una colonna è dipendente dalle altre. Vuol dire anche che la trasformazione porta a uno spazio con dimensione inveriore.

Se invece il valore del determinante è negativo?. Cosa possiamo dire una area non può essere negativa, ma se pensiamo che una trasformazione può invertire l'ordine degli assi questo da origine ad un valore negativo del determinante.

Nello spazio tridimensionale questo si trasforma nel volume del cubo.
Per uno spazio bidimensionale il determinante si calcola in questo modo:

$$
det \left(\begin{bmatrix} a & b \\ c & d \end{bmatrix} \right) = ad - bc
$$

Ora sempre utilizzando il concetto di trasformazione per introdurre dei nuovi concetti:

* Matrice inversa
* Spazio delle colonne
* Rango

Prima di iniziare questo dobbiamo unire al concetto di trasformazione quello dei sistemi lineari visti prima.

se prendiamo il sistema visto prima possiamo pensare che la matrice A rappresenta una trasformazione, mentre $\vec{v}$ rappresenta il risultato.
Sotto questa premessa ora il sistema lo possiamo vedere come quale è il vettore [x,y,z] che applicato alla matrice di trasformazione A da come risultato $\vec{v}$ ?

## Matrice inversa

Con l'informazione di prima possiamo ora ragionare sul significato della matrice inversa, questa matrice inverte la trasformazione che abbiamo appena fatto riportando $\vec{v}$ ai suoi valori originali.

Percio se eseguiamo la seguente operazione $A^{-1}A$ otterremo la matrice identità che rappresenta una matrice in cui non c'è nessuna trasformazione.

Per trovare la soluzione al sistema basta dunque moltiplicare $A^{-1}$ per $\vec{v}$:

$$A^{-1}A\vec{x} = A^{1}\vec{v} \rightarrow \vec{x} = A^{1}\vec{v}$$

ovviamente questo ragionamento vale se $det(A) \neq 0$ se siamo nel caso in cui $det(A) = 0$ non esisterà la matrice inversa in quanto un singolo punto di uno spazio vettoriale può corrispondere ad una infinità di punti su uno spazio vettoriale più grande.

Questo ci apre la strada ad un nuovo concetto.

## Rango di una matrice

Il rango rappresenta il numero delle dimensioni del vettore dopo la trasformazione lineare 

## Spazio nullo

Lo spazio nullo rappresenta la soluzione del seguente sistema (per un sistema a due dimensioni):

$$A\vec{x} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$

Praticamente rappresenta l'insieme di tutti i vettori tale per cui la trasformazione mappa il risultato in 0 su tutti gli assi.




## Prodotto vettoriale

Prendiamo i soliti vettori $\vec{w}$ su $\vec{v}$ e pensiamo all'area del parallelogramma che questi generano, il prodotto vettoriale di questi due vettori rappresenta l'area del parallelogramma
Dobbiamo considerare anche l'orientamento dei vettori, se $\vec{v}$ è sulla sinistra di  $\vec{w}$ come nella immagine sotto, il prodotto vettoriale sarà positivo altrimenti tale protto sarà negativo.

![cross](images/cross1.png)

sappiamo anche della seguente ugualianza 

$$\vec{v} \times \vec{w} = -\vec{w} \times \vec{v}$$

Sappiamo anche che se prendiamo i vettori unitari $\vec{i}$ e $\vec{j}$ il prodotto vettoriale $\vec{i} \times \vec{j}$ avrà risultato 1

![cross](images/cross2.png)


come abbiamo visto questa area è rappresentata anche dal determinante della matrice, ora se i vettori sono ortogonali l'area sarà la più grande possibile e continuerà a diminuire fino a che i vettori non diventano paralleli.

Anche qui se moltiplichiamo un vettore per una costante l'area aumenta per quella costante.
Ma il prodotto vettoriale da come risultato un vettore ... e la lunghezza del vettore sarà rappresentata dall'area, ma il verso?
Abbiamo detto che il vettore risultanto sarà perpendicolare al piano generato dai due vettori, ma come definiamo il verso.

Qui ci viene il aiuto la regola della mano destra, per calcolare il vettore risultante si può usare la formula qui sotto:

![cross](images/cross3.png)


Proviamo a vedere il perchè della formula se scriviamo i vettori nel seguente modo:

$$\begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix} \times \begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix} = det \left(
\begin{bmatrix} i & v_1 & w_1 \\ j & v_2 & w_2 \\ k & v_3 & w_3 \end{bmatrix} \right)$$

Il determinante risultate è:

$$i(v_2w_3 - v_3w_2) + j(v_3w_1 - v_1w_3) + k(v_1w_2 - v_2w_1)$$

Ed ecco ricavata la formula

## Cambio Di base

![base](images/base.png)


Abbiamo usato implicitamente un sistema di coordinate, nel nostro caso i vettori $i$ e $j$ rappresentano una base.
Ma se avessimo usato un insieme di vettori differente ?

Supponiamo di utilizzare una nuova base che chiameremo  $b_1$ e $b_2$ 

![base](images/base1.png)

Ovviamente nel nuovo sistema di coordinate i vettori $b_1$ e $b_2$ sono unitari.
Come possiamo portare i vettori da una base all'altra?

Supponiamo di avere un vettore $-1b_1 + 2b_2$ quale sarà il suo corrispondente nella base di partenza?
Questa operazione si chiama cambio di base, ma è una cosa che abbiamo già visto si tratta di una trasformazione lineare. 

La matrice che rappresenta la nostra base nel vecchio sistema è la matrice di trasformazione.
